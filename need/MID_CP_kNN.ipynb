```python
from PIL import Image
import os

input_directory = 'CIFAR-10-images/test'
output_directory = 'Data/test_grayscale'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for folder in os.listdir(input_directory):
    folder_path = os.path.join(input_directory, folder)
    output_folder = os.path.join(output_directory, folder)

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file in os.listdir(folder_path):
        if file.endswith(".jpg"):
            input_path = os.path.join(folder_path, file)
            output_path = os.path.join(output_folder, file)
            img = Image.open(input_path)
            img_gray = img.convert('L')
            img_gray.save(output_path)

```

```python
import os
from PIL import Image
import random
import matplotlib.pyplot as plt

input_dir = 'Data/test_grayscale'

subdirs = [subdir for subdir in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, subdir))]

for subdir in subdirs:
    subdir_path = os.path.join(input_dir, subdir)
    image_files = [file for file in os.listdir(subdir_path) if file.endswith(".jpg")]

    if image_files:
        random_image = random.choice(image_files)
        image_path = os.path.join(subdir_path, random_image)

        img = Image.open(image_path)
        plt.figure()
        plt.imshow(img, cmap='gray')
        plt.title(f"Image from {subdir}")
        plt.axis('off')

plt.show()
```
